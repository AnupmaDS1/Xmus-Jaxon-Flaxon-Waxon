```python
# import the crime data files first which are in github releases folder 
import requests

url = "https://github.com/AnupmaDS1/Xmus-Jaxon-Flaxon-Waxon/releases/download/v1.1-data/WYP2024_2025.zip"
r = requests.get(url)

print(r.status_code)
print(r.headers.get("content-type"))

with open("Crimedata", "wb") as f:
    f.write(r.content)
```
